```python
import pandas as pd
import numpy as np
```

```python
# Reading in the weather data
data = pd.read_csv("../../data/processed/clean_weather_data.csv")
```

### Basic Statistics of the Data

```python
# checking for max, min, std, and variance
for label in data.select_dtypes(include='number').columns:
    print(f"{label} - std: {data[label].std()} var: {data[label].var()} min: {data[label].min()} max: {data[label].max()}")
```

```python
import matplotlib.pyplot as plt

# Plot each numerical feature
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Define the columns to plot
features = ['wind_speed', 'mean_temp', 'global_radiation', 'percipitation', 'pressure', 'minimum_visibility', 'humidity']
for ax, feature in zip(axes.flatten(), features):
    data[feature].plot(ax=ax, title=feature)
    ax.set_xlabel('date')
    ax.set_ylabel(feature)

plt.tight_layout()
plt.show()
```
